```python
import numpy as np
import pandas as pd
from scipy import stats

# ... Your previous code to load and clean the dataset

data = df.select_dtypes(include=np.number) # Select only numeric columns

# Five Number Summary
five_num_summary = {
    'Minimum': np.min(data),
    'Q1 (25th percentile)': np.percentile(data, 25),
    'Median (Q2 - 50th percentile)': np.median(data),
    'Q3 (75th percentile)': np.percentile(data, 75),
    'Maximum': np.max(data)
}

# Mode
# For mode calculation, you might want to consider doing it per column
# as the mode might not be meaningful across all numeric columns
mode_values = {col: stats.mode(data[col]) for col in data.columns}

# Midrange
midrange = (np.min(data) + np.max(data)) / 2

# Outlier Detection using Quartile method
Q1 = five_num_summary['Q1 (25th percentile)']
Q3 = five_num_summary['Q3 (75th percentile)']
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]

# Printing the results
print("Five Number Summary:")
for key, value in five_num_summary.items():
    print(f"{key}: {value}")

print("\nMode (per column):")
for col, mode in mode_values.items():
    if isinstance(mode.mode, np.ndarray):  # Check if mode is an array
        print(f"Column {col}: Mode: {mode.mode[0]}, Count: {mode.count[0]}")
    else:
        print(f"Column {col}: Mode: {mode.mode}, Count: {mode.count}")  # Handle scalar mode

print("\nMidrange:")
print(midrange)

print("\nOutliers Detected (consider checking per column):")
print(outliers)

# Additional statistics (for comparison)
mean = np.mean(data)
std_dev = np.std(data)
print("\nAdditional Statistics:")
print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
```
